```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a hypothetical dataset
np.random.seed(0)
data = pd.DataFrame({
    'Age': np.random.randint(1, 100, size=1000)  # 1000 random ages between 1 and 100
})

# Display the first few rows of the dataset
print(data.head())

# Create a histogram to visualize the distribution of ages
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
ggggg
```

```python
# Create a hypothetical dataset for gender
data = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female', 'Other'], size=1000, p=[0.48, 0.48, 0.04])
})

# Display the first few rows of the dataset
print(data.head())

# Create a bar chart to visualize the distribution of genders
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
